### **Product Analysis**

**Probability of a Male Customer Buying a KP781 Treadmill**
- We use conditional probability
   P(KP781|Male) = Number of Males Who Purchased KP781 / Total number of Male Customers

```python
df[df['Gender'] == 'Male']['Product'].value_counts()
```

```python
df[df['Gender'] == 'Male']['Product'].value_counts(normalize=True)
```

```python
df[df['Gender'] == 'Male']['Product'].value_counts(normalize=True).get('KP781', 0)
```

```python
# Conditional Probability: P(KP781 | Male)
prob_KP781_given_male = df[df['Gender'] == 'Male']['Product'].value_counts(normalize=True).get('KP781', 0)
print("P(KP781 | Male):", prob_KP781_given_male)
```

```python
# Marginal Probability of purchasing each treadmill
marginal_prob = df['Product'].value_counts(normalize=True)
print("Marginal Probability:\n", marginal_prob)

# Conditional Probability: P(KP781 | Male)
prob_KP781_given_male = df[df['Gender'] == 'Male']['Product'].value_counts(normalize=True).get('KP781', 0)
print("P(KP781 | Male):", prob_KP781_given_male)

# Conditional Probability: P(KP481 | Income > 60K)
prob_KP481_given_income = df[df['Income'] > 60000]['Product'].value_counts(normalize=True).get('KP481', 0)
print("P(KP481 | Income > 60K):", prob_KP481_given_income)
```

#Observation:
- **Marginal Probabilities:**
  - 44.44 Percent of customers bought the KP281 Treadmill
  - 33.33 Percent of customers bought the KP481 Treadmill
  - 22.22 Percent of customers bought the KP781 Treadmill
- **Conditional Probabilities**
  - 31.73 Percent of male customers bought the KP781 Treadmill.
  - 16.67 Percent of customers earning more than 60K Purchased the KP481 Treadmill.

###**HEATMAP FOR CORRELATION MATRIX**

```python
# Visualize correlation between numerical features like
# (Age, Income, Miles, Usage, and Fitness)
plt.figure(figsize = (8,6))
correlation_matrix = df[['Age', 'Income', 'Miles', 'Usage', 'Fitness']].corr()
sns.heatmap(correlation_matrix, annot = True, cmap = 'coolwarm')
plt.title('Correlation Matrix')
plt.show()
```

#Observation:
- **Miles & Fitness:** People who run/walk more miles tend to have higher fitness levels.
- **Miles & Usage:** More frequent use of fitness equipment leads to covering more miles.
- **Usage & Fitness:** Higher usage is linked to better fitness, reinforcing the effectiveness of regular workouts.
- **Income & Miles/Fitness:** Higher income individuals tend to cover more miles and have better fitness.
