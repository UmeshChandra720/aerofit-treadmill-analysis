- **PLOT DISTRIBUTION FOR AGE, INCOME, MILES**

- **AGE DISTRIBUTION OF CUSTOMERS**

```python
# Visaulize the distribution of customer ages
plt.figure(figsize=(10,4))
sns.histplot(df['Age'], bins=20, kde=True)
plt.title('Distribution of Customer Ages')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()
```

- **INCOME DISTRIBUTION OF CUSTOMERS**

```python
# Display the Income distribution of customers
plt.figure(figsize=(10,4))
sns.histplot(df['Income'], bins=20, kde=True)
plt.title('Distribution of Customer Income')
plt.xlabel('Income')
plt.ylabel('Frequency')
plt.show()
```

- **USAGE FREQUENCY OF TREADMILL**

```python
# Distribution of usage
plt.figure(figsize=(8,5))
sns.histplot(df['Usage'], bins=10, kde=False, color = 'orange')
plt.title('Distribution of Usage')
plt.xlabel('Usage')
plt.ylabel('Frequency')
plt.show()
```

- **SPREAD OF CUSTOMER AGES**

```python
# Detect outliers and spread of customer ages
plt.figure(figsize=(10,4))
sns.boxplot(x=df['Age'])
plt.title('Outliers on Customer Ages')
plt.xlabel('Age')
plt.show()
```

- **INCOME SPREAD ACROSS CUSTOMERS**

```python
# Identify outliers and Income spread across customers
plt.figure(figsize = (10,4))
sns.boxplot(data = df, x = "Income")
plt.title(' Income spread across Customer')
plt.xlabel('Income')
plt.xticks(rotation = 0)
plt.show()
```

- **MALE VS. FEMALE CUSTOMERS**

```python
# Show the frequency of male vs. female customers.
plt.figure(figsize = (3,4))
sns.countplot(data = df, x = 'Gender')
plt.title('Male vs. Female Customers')
plt.xlabel('Gender')
plt.show()
```

- **FREQUENCY OF PRODUCT PURCHASED**

```python
# frquency of each treadmill model
plt.figure(figsize = (6,4))
sns.countplot(data = df, x = 'Product')
plt.title('Frequency of Product Purchased')
plt.xlabel('Product')
plt.show()
```

- **BIVARIATE ANALYSIS**

- **GENDER VS. MARITALSTATUS**

```python
# Gender vs. MaritalStatus
plt.figure(figsize = (6,6))
sns.countplot(data = df, x = 'Gender', hue = 'MaritalStatus')
plt.title('Gender vs. MaritalStatus')
plt.xlabel('Gender')
plt.ylabel('Count')
plt.show()
```

- **FITNESS LEVEL BY GENDER**

```python
# Fitness rating differ between genders
plt.figure(figsize = (6,4))
sns.boxplot(data = df, y = 'Fitness', x = 'Gender')
plt.title('Fitness rating differ between genders')
plt.xlabel('Gender')
plt.ylabel('Fitness')
plt.show()
```

- **AGE VS. PRODUCT PURCHASED**

```python
# Anlayse the relationship between customer age
# and product purchased
plt.figure(figsize = (6,4))
sns.boxplot(data = df, x = 'Product', y = 'Age')
plt.title('Age vs. Product Purchased')
plt.ylabel('Age')
plt.xlabel('Product')
plt.show()
```

- **INCOME VS. PRODUCT PURCHASED**

```python
# Relationship between Income and product purchased
plt.figure(figsize = (6,4))
sns.boxplot(data = df, x = 'Product', y = 'Income')
plt.title('Income vs. Product Purchased')
plt.ylabel('Income')
plt.xlabel('Product')
plt.show()
```

- **HEATMAP FOR CORRELATION MATRIX**

```python
# Visualize correlation between numerical features like
# (Age, Income, Miles, Usage, and Fitness)
plt.figure(figsize = (8,6))
correlation_matrix = df[['Age', 'Income', 'Miles', 'Usage', 'Fitness']].corr()
sns.heatmap(correlation_matrix, annot = True, cmap = 'coolwarm')
plt.title('Correlation Matrix')
plt.show()
```

- **PAIRPLOT FOR NUMERICAL VARIABLES**

```python
# Explore relationship between pairs of numerical variables
# Explore relationship between pairs of numerical variables
# and their interaction with the product purchased
sns.pairplot(df[['Age', 'Income', 'Miles', 'Usage', 'Fitness', 'Product']],
             hue = 'Product', palette = 'husl')
plt.suptitle('Pairplot of Numerical Variables', y=1.02)
plt.show()
```
